```python
import matplotlib.pyplot as plt
import numpy as np

# Générer des données pour une relation linéaire positive
x = np.linspace(0, 10, 100)
y = 2 * x + np.random.normal(0, 1, 100)

plt.scatter(x, y)
plt.title("Relation Linéaire Positive")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

```python
# Générer des données pour une relation linéaire négative
x = np.linspace(0, 10, 100)
y = -2 * x + np.random.normal(0, 1, 100)

plt.scatter(x, y)
plt.title("Relation Linéaire Négative")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```
